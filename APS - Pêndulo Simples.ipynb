```python
import matplotlib.pyplot as plt
import math
from scipy.integrate import odeint
import numpy as np

g = 9.8 #aceleração da gravidade
l = 1.5 #comprimento da haste
teta = math.radians(10) #ângulo inicial (teta)
tempo = np.arange(0, 15, 0.001)



def Pendulo(Y, tempo):
    omega = Y[0] #velocidade angular
    teta = Y[1] 
    domegadt = -g*math.sin(teta)/l #aceleração angular
    dtetadt = omega #velocidade angular

    return [domegadt, dtetadt]


angulo_pendulo = odeint(Pendulo, [0, teta], tempo)


teta_radians = angulo_pendulo[:,1]


x = []
y = []
T_exato = 0
T_formula = 2*math.pi*math.sqrt(l/g) #fórmula do período

for i in range(len(teta_radians)):
    x.append(l*math.sin(teta_radians[i])) #preenche a lista de x (trajetória em x)
    y.append(-l*math.cos(teta_radians[i])) #preenche a lista de y (trajetória em y)



for t in range(0, len(teta_radians)):
    teta_radians[t]=round(teta_radians[t], 5)
    
     
for s in range(0, len(x)):
    x[s]=round(x[s],6)

    
for n in range(1, len(x)): #Encontra o período exato
    if x[n]==x[0]:
        T_exato=tempo[n]
        break
        

        
        
print("O período exato do pêndulo, considerando-se uma amplitude de {0}° é de {1}s." .format((round(math.degrees(teta))), T_exato))
print("O período do pêndulo, calculado segundo a fórmula, é de {0}s." .format(T_formula))
plt.plot(x,y, color='fuchsia')
plt.axis('equal')
plt.grid()
plt.title('Trajetória de um pêndulo com ângulo de amplitude {0}°'.format(round(math.degrees(teta))))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(tempo, x, color='orange')
plt.axis('equal')
plt.grid()
plt.title('Trajetória em X (ângulo inicial de amplitude {0}°)'.format(round(math.degrees(teta))))
plt.xlabel('tempo')
plt.ylabel('x')
plt.show()
```

```python
plt.plot(tempo, y, color='orange')
plt.axis('equal')
plt.grid()
plt.title('Trajetória em y (ângulo inicial de amplitude {0}°)'.format(round(math.degrees(teta))))
plt.xlabel('tempo')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(tempo, teta_radians, color='orange')
plt.axis('equal')
plt.grid()
plt.title('Oscilação do ângulo teta (ângulo inicial de amplitude {0}°)'.format(round(math.degrees(teta))))
plt.xlabel('tempo')
plt.ylabel('teta')
plt.show()
```
